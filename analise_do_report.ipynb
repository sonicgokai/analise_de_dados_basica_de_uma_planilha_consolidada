## Analisando a margem de  lucro


```python
df.head()
```

### Agrupando dados

#### por segmento

```python
colunas = ['valor_toital_desconto','custo_total','lucro']
df_segmento = df.groupby('segmento')[colunas].sum().round(2)
```

```python
type(df_segmento)
```

```python
df_segmento.head()
```

```python
df_segmento['margem_lucro'] = round(df_segmento['lucro'] / df_segmento['valor_toital_desconto'] * 100,2)
```

```python
df_segmento
```

#### por produto

```python
colunas = ['valor_toital_desconto','custo_total','lucro']
df_produto = df.groupby('produto')[colunas].sum().round(2)
```

```python
df_produto.head()
```

```python
df_produto['margem_lucro'] = round(df_produto['lucro'] / df_produto['valor_toital_desconto'] * 100,2)
```

```python
df_produto
```

#### por pais

```python
colunas = ['valor_toital_desconto','custo_total','lucro']
df_pais = df.groupby('pais')[colunas].sum().round(2)
df_pais['margem_lucro'] = round(df_pais['lucro'] / df_pais['valor_toital_desconto'] * 100,2)
df_pais
```
